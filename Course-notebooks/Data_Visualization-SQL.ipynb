```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

# Hands-on Lab


## Visualizing distribution of data


```python
import matplotlib.pyplot as plt
import seaborn as sns
```

### Histograms


Plot a histogram of `ConvertedComp.`


```python
QUERY = """
SELECT * FROM master
"""
df = pd.read_sql_query(QUERY,conn)
df.head()
```

```python
# your code goes here

df['ConvertedComp'].plot(kind='hist', figsize=(9, 12))
plt.ticklabel_format(style='plain')
plt.title("Annual Salary in USD of the Respondents", size =15)
plt.xlabel("Compensation ($US)", size = 12)
plt.ylabel("Frequency", size = 12)
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here

df['Age'].plot(kind='box', figsize=(9, 12))
plt.title("Age of the Respondents", size = 15)
plt.ylabel("Value")
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here

fig = sns.scatterplot(x='Age', y='WorkWeekHrs', data=df)
fig = plt.gcf()
fig.set_size_inches(9, 9)
plt.title("Comparision Between Age and Work Week Hours", size = 15)
plt.ylabel("Work Week Hours", size = 12)
plt.xlabel("Age", size = 12)
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here

sns.scatterplot(data=df, x='WorkWeekHrs', y='CodeRevHrs', size='Age', hue='Age', sizes=(10, 300))
fig = plt.gcf()
fig.set_size_inches(9, 10)
plt.title('Work Week Hours and Code Rev Hours By Age', size=15)
plt.xlabel('Work Week Hours', size=12)
plt.ylabel('Code Rev Hours', size=15)

plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here

QUERY = """ 
SELECT DatabaseDesireNextYear, COUNT(*) as Total 
from DatabaseDesireNextYear 
group by DatabaseDesireNextYear 
order by count(DatabaseDesireNextYear) DESC LIMIT 5 
""" 

df2=pd.read_sql_query(QUERY,conn)
df2.set_index("DatabaseDesireNextYear", inplace=True) 
colors_list=['lightgreen', 'gold', 'lightcoral', 'yellowgreen', 'lightskyblue']
df2['Total'].plot(kind='pie', figsize=(9.5, 6), autopct='%1.1f%%', labels=None, startangle=90, colors=colors_list, shadow=True, pctdistance=1.16)  

plt.legend(labels=df2.index) # add legend
plt.title('Top 5 Databases Respondents Wish To Learn', size = 15) 
plt.axis('equal')
plt.ylabel('Percentage', size = 12)

plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
QUERY = """
SELECT WorkWeekHrs, CodeRevHrs, Age FROM master
WHERE Age BETWEEN 30 AND 35
"""
df3 = pd.read_sql_query(QUERY,conn)
df4 = df3.groupby('Age').median()
df4.plot(kind='bar', figsize=(9, 9), stacked=True, ylim=(0, 50))
plt.title('Median Work Week Hours and Code Rev Hours for Age Group 30 to 35', size = 15)
plt.xlabel('Age', size = 12)
plt.ylabel('Median', size = 12)
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
QUERY = """
SELECT ConvertedComp, Age FROM master
WHERE Age BETWEEN 45 AND 60
"""
df5 = pd.read_sql_query(QUERY,conn)
df6 = df5.groupby('Age').median()

df6.plot(kind='line', figsize=(9, 8))

plt.title('Median Annual Salary in USD for Age Group 45 to 60', size = 15)
plt.xlabel('Age', size = 12)
plt.ylabel('Annual Salary', size = 12)
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
#df['MainBranch'].value_counts()

QUERY = """
SELECT MainBranch, COUNT(*) as Total
from master
group by MainBranch
"""

df7=pd.read_sql_query(QUERY,conn)
df7.set_index("MainBranch", inplace=True) 
fig = df7.plot(kind='barh', figsize=(8.2,10), color='lightskyblue', legend=False)

y_labels = ["Developer by \n Profession ", "Not a Developer \n but I write code \n as part of my job "]
fig.set_yticklabels(y_labels) #change de labels at the end of each bar
plt.title('Main Branch of the Respondents', size = 15)
plt.xlabel('Frequency', size = 12)
plt.ylabel('Main Branch', size = 10)
plt.show()
```
